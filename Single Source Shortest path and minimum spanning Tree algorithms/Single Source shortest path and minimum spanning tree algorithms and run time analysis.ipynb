```python
from collections import defaultdict
import sys
import time
Dijkstra_times = []

class Graph:
    def __init__(self, directed = False, vertices = None):
        self.directed = directed
        self.graph = defaultdict(list)
        self.vertices = vertices
        
    def add_edge(self, frm, to, cost):
        self.graph[frm].append([to, cost])

        if self.directed is False:
            self.graph[to].append([frm, cost])
        elif self.directed is True:
            self.graph[to] = self.graph[to]   
            
    def get_edges(self):
        print(self.graph.keys())
        print(self.graph.values())
        
        
    def findMinDist(self, distance, visit):
        min_dist = float('inf')
        index = -1
        for key in self.graph.keys():
            if visit[key] is False and distance[key] < min_dist:
                min_dist = distance[key]
                index = key
        return index
    
    def path(self,root, key):
        if root[key] is None:
            return
        self.path(root, root[key])
        print(chr(key+65),end=" ")
    
    def solution(self, distance, root, source):
        print('{} \t\t{} \t {}'.format('Vertex', 'Distance', 'Path'))

        for key in self.graph.keys():
            if key == source:
                continue
            if  distance[key]==float("inf"):
                continue
            print('{} -> {} \t\t {} \t\t {}'.format(chr(source+65), chr(key+65), distance[key], chr(source+65)), end=' ')
            temp = []
            self.path(root, key)
            print()
            
    def dijkstra(self, source):
        visited = {i: False for i in self.graph}
        distance = {i: float('inf') for i in self.graph}
        root = {i: None for i in self.graph}

        distance[source] = 0

        # find shortest path for all vertices
        for i in range(len(self.graph) - 1):
            start = self.findMinDist(distance, visited)
            visited[start] = True
            for vertex, cost in self.graph[start]:

                if visited[vertex] is False and distance[start] + cost < distance[vertex]:
                    distance[vertex] = distance[start] + cost
                    root[vertex] = start
        return root, distance


if __name__ == '__main__':
    directed = False
    N_inputs = 0
    
    while N_inputs < 6:
        filepath = "graph" + str(N_inputs) + ".txt"
        inputFile = open(filepath,'r')
        N_line = 0
        source = sys.maxsize
        print('\033[1m'+"Single source Shortest paths for"+'\033[0m'+" {} is: ".format(filepath))
        print("=======================================================================")
        
        for line in inputFile.readlines():
            temp = line.split()
            if N_line == 0:
                print("Number of Vertices provided: {}".format(int(temp[0])))
                print("Number of Edges provided: {}".format(int(temp[1])))
                d_ud = temp[2]
                if d_ud == "D":
                    directed = True
                    print("The Graph is: DIRECTED")
                else:
                    directed = False
                    print("The Graph is: UN DIRECTED")
                G = Graph(directed)
            elif len(temp) == 1:
                source = ord(temp[0])-65
            else:
                G.add_edge(ord(temp[0])-65, ord(temp[1])-65, int(temp[2]))
            
            N_line += 1
        
        print("Start or Source provided: {}".format(chr(source+65)))
        st = time.time()
        root, distance = G.dijkstra(source)
        
        #print(parent, dist)
        print("******************************************")
        print("Dijkstra's Algorithm results:")
        print("******************************************")
        G.solution(distance, root, source)
        N_inputs += 1
        print("***********************************************************************")
        print('\033[1m'+"Time taken for this input to execute is :" +'\033[0m', (time.time() - st)*1000,"\n")
        print("***********************************************************************\n")
```

# Problem 2: 

## Minimum Spanning Tree

# Kruskal's Algorithm

```python
class Graph:

    def __init__(self, vertices):
        self.vertices = vertices
        self.graph = []


    def add_edge(self, root, vertex, cost):
        self.graph.append([root, vertex, cost])

    def get_root_node(self, root_node, v):
        if root_node[v] == v:
            return v
        return self.get_root_node(root_node, root_node[v])

    def union_of_sets(self, root_node, rank, a, b):
        root_a = self.get_root_node(root_node, a)
        root_b = self.get_root_node(root_node, b)

        if rank[root_a] < rank[root_b]:
            root_node[root_a] = root_b
        elif rank[root_a] > rank[root_b]:
            root_node[root_b] = root_a

        else:
            root_node[root_b] = root_a
            rank[root_a] += 1

    def Kruskal(self):
        result = []
        idx_e = 0
        idx_r = 0
        self.graph = sorted(self.graph, key=lambda item: item[2])
        parent = []
        rank = []
        for node in range(self.vertices):
            parent.append(node)
            rank.append(0)
        while idx_e < self.vertices - 1:

            root, vertex, cost = self.graph[idx_r]
            idx_r += 1
            s1 = self.get_root_node(parent, root)
            s2 = self.get_root_node(parent, vertex)
            if s1 != s2:
                idx_e += 1
                result.append([root, vertex, cost])
                self.union_of_sets(parent, rank, s1, s2)

        print('\033[1m' + "Edge Selected \t\t Weight" + '\033[0m')
        print()
        total_cost=0
        for a, b, c in result:
            print(chr(a + 65), "->", chr(b + 65), "\t\t\t  ", c)
            total_cost += c
        print("***************************************************************************************")
        print('\033[1m'+"The total Cost for Minimum Spanning Tree is", total_cost,'\033[0m')   
        

if __name__ == '__main__':
    n_inputs = 0
    print()
    while (n_inputs < 6):
        print("=======================================================================================")
        filepath = "graph" + str(n_inputs) + ".txt"
        inputFile = open(filepath, "r")
        print('\033[1m'+"The Minimum Spanning Tree using Kruskal's Algorithm for"+'\033[0m'+" {} :" .format(filepath))
        print("=======================================================================================")    

        n_line = 0
        for line in inputFile.readlines():
            temp = line.split()
            if n_line == 0:
                no_of_vertices = int(temp[0])
                graph = Graph(no_of_vertices)
            elif len(temp) == 1:
                pass
            else:
                graph.add_edge(ord(temp[0]) - 65, ord(temp[1]) - 65, int(temp[2]))
            n_line = n_line + 1
    
        print("**************************************")
        print("Kruskal's Algorithm Results")
        print("**************************************")
        st = time.time()
        graph.Kruskal()
        n_inputs += 1
        print("***************************************************************************************")
        print('\033[1m'+"Time taken for this input to execute is :" +'\033[0m', (time.time() - st)*1000)
        print("***************************************************************************************\n")

```
